## 1.2 Example

### Load data

open the data file 'm-gm3dx7508.txt' as data

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf

# Load data
data = pd.read_csv("m-gm3dx7508.txt", delim_whitespace=True, index_col = 0)
```

```python
data
```

### Extract GM stock returns (it's in the second column)

```python
gm = data.iloc[:,1]
gm
```

### Create a time series object (Pandas Series with a DateTime index)

The data is monthly and starts from January 1975

```python
gm1 = pd.Series(gm.values, index=pd.date_range(start='1975-01-01', periods=len(gm), freq='M'))
```

```python
gm1
```

### Plotting the data

```python
plt.figure(figsize=(12, 8))
plt.subplot(2, 1, 1)
plt.plot(gm, linestyle='-', marker='None')
plt.title("GM Stock Returns (Original)")

plt.subplot(2, 1, 2)
plt.plot(gm1, linestyle='-', marker='None')
plt.title("GM Stock Returns (Time Series)")

plt.tight_layout()
plt.show()
```

### Plotting ACF

```python
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plot_acf(gm, lags=24, ax=plt.gca())
plt.title("Autocorrelation Function (Original)")

plt.subplot(2, 1, 2)
plot_acf(gm1, lags=24, ax=plt.gca())
plt.title("Autocorrelation Function (Time Series)")

plt.tight_layout()
plt.show()
```

## Question 2.3. 
Consider the monthly U.S. unemployment rate from January 1948 to March 2009 in the file **[m-unrate.txt](https://faculty.chicagobooth.edu/-/media/faculty/ruey-s-tsay/teaching/fts3/m-unrate.txt)**. The data are seasonally adjusted and obtained from the Federal Reserve Bank of St Louis. Build a time series
model for the series and use the model to forecast the unemployment rate for the April, May, June, and July of 2009. In addition, does the fitted model imply the existence of business cycles? Why? (Note that there are more than one model fits the data well. You only need an adequate model.

## Solution
To build a time series model for the U.S. unemployment rate and forecast future values, you would typically follow these steps:

1. Load the data into a pandas DataFrame.
2. Plot the data to visualize any trends, seasonality, or patterns.
3. Decide on an appropriate time series model (e.g., ARIMA, Seasonal ARIMA, etc.).
4. Fit the model to the data.
5. Use the fitted model to make forecasts.
6. Examine the fitted model for any implications of business cycles.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Step 1: Load the data
data = pd.read_csv('./data/m-unrate.txt', delim_whitespace=True, header=0)
data
```

```python
#unemployment_rate = data['Rate']  # replace 'UnemploymentRate' with the correct column name
data['Date'] = pd.to_datetime(data[['Year', 'Mon', 'Day']].rename(columns = {"Mon" : "Month"}))
data.set_index('Date', inplace=True)
data.drop(['Year','Mon','Day'], axis =1, inplace = True)
data
```

```python
# Step 2: Plot the data
plt.plot(data.index, data['Rate'], label='Unemployment Rate')
plt.title('Monthly U.S. Unemployment Rate')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate (%)')
plt.legend()
plt.show()
```

```python
# Step 3: Decompose the series (optional, helps in identifying seasonal patterns)
decomposition = seasonal_decompose(data, model='additive')
```

```python
decomposition.plot()
plt.show()
```

```python
# Step 4: Fit a Seasonal ARIMA model
# Note: The order and seasonal_order should be selected based on model diagnostics (like AIC)
model = SARIMAX(data, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
results = model.fit()
```

```python
# Assuming 'results' is the fitted SARIMAX model from the previous steps
# Forecast the unemployment rate for the next four months
forecast_results = results.get_forecast(steps=4)
forecast_mean = forecast_results.predicted_mean
forecast_conf_int = forecast_results.conf_int()

# Print the forecasted values
print(forecast_mean)

# Print the confidence intervals for the forecasts
print(forecast_conf_int)

# You can also plot the forecasted values alongside the original series
plt.figure(figsize=(10, 5))
plt.plot(data.index, data, label='Observed')
plt.plot(forecast_mean.index, forecast_mean, label='Forecast', color='red')
plt.fill_between(forecast_conf_int.index,
                 forecast_conf_int.iloc[:, 0],
                 forecast_conf_int.iloc[:, 1], color='pink', alpha=0.3)
plt.title('Forecast of the Unemployment Rate')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate')
plt.legend()
plt.show()
```

```python
# Business cycle analysis (simplistic approach)
# Check if the fitted model's residuals show any cyclical patterns
residuals = results.resid
plt.plot(residuals)
plt.title('Residuals of the Fitted Model')
plt.show()
```

```python
# Autocorrelation check
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(residuals, lags=50)
plt.title('ACF of Residuals')
plt.show()
```

```python
plot_pacf(residuals, lags=50)
plt.title('PACF of Residuals')
plt.show()
```
